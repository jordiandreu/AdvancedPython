# Exercice 1

```python
class Wrapper:
    def __init__(self, wrapped, non_allowed=None):
        self._wrapped = wrapped
        self._non_allowed = non_allowed
        
    def __getattr__(self, attr):
        if attr not in self._non_allowed:
            return getattr(self._wrapped, attr)
        raise AttributeError(attr)
    def __len__(self):
        return self._wrapped.__len__()
```

```python
w = Wrapper([], 'append')
```

```python
w.append(1)
```

```python
len(w)
```

```python
w.extend(range(5))
```

```python
len(w)
```

# Exercice 2

```python
class Calculator:

    def __init__(self, compute=None):
        
        self.compute = self.__compute
        
        if hasattr(compute, '__call__'):
            self.compute = compute
            
    @staticmethod
    def __compute():
        return 42
            
```

```python
def func():
    return 1
```

```python
c = Calculator(func)
```

```python
c.compute()
```

```python
c = Calculator()
```

```python
c.compute()
```

```python
def factory(arg):
    def _func(x):
        return x * arg
    return _func
```

```python
f10 = factory(10)
f10(5)
```

```python
f50 = factory(50)
f50(5)
```

## Appendix

```python
def make_func(func=None):
    
    if callable(func):
        func = func
    else:
        func = lambda x, y: x + y
    def proxy(*args, **kwargs):
        proxy.counter += 1
        return func(*args, **kwargs)

    proxy.counter = 0
    
    return proxy
```

```python
class Func:
    def __init__(self, func=None):
        if callable(func):
            self.func = func
        else:
            self.func = lambda x, y: x+y
        self.counter = 0
    def __call__(self, *args, **kwargs):
        self.counter += 1
        return self.func(*args, **kwargs)
```

```python
func = make_func()
```

```python
func.counter
```

```python
func2 = Func()
```

```python
func2.counter
```
